```python
from bs4 import BeautifulSoup
import requests
import pprint

import re
import pandas as pd
```

### Return all my album reviews

```python
pages = []

for i in range(1, 100):
    url = 'https://pitchfork.com/reviews/albums/?page=' + str(i)
    pages.append(url)
    soup_list = []
    
for item in pages:
    response = requests.get(item)
    page = response.text
    soup_list.append(BeautifulSoup(page, "lxml"))
```

### This works to get rid of those links

```python
article_link_list = []

for sl in soup_list:
    r = re.compile("https://pitchfork.com/reviews/albums/\w+")
    article_link_list.append([x for x in (list(filter(r.match,['https://pitchfork.com' + x['href'] for x in sl.find_all('a')])))\
          if x!= 'https://pitchfork.com/reviews/albums/popular/'])
```

### This reads back the contents of all the links within the links

```python
# Define a function contents_in_link that returns the contents of the links within the All_review_links1
# variable and store as a variable named link_contents
def contents_in_link():
    contents_list = []
    for art_list in article_link_list: # for each list within the list
        for arl in art_list: # for each link within the list of lists
            url = arl
            response = requests.get(url)
            page = response.text
            soup = BeautifulSoup(page,"lxml")
            contents_list.append(soup)    
    return contents_list

link_contents = contents_in_link()
```

### Pull back author

```python
author_list = []

for lc in link_contents:
    try: 
        author_list.append(lc.find(class_='article-meta article-meta--reviews').find('a').text)
    except:
        author_list.append('No author')
        
author_list
```

### Pull back score

```python
review_score = []

for lc in link_contents:
    try:
        review_score.append(lc.find(class_='single-album-tombstone').find(class_='score').text)
    except:
        review_score.append('No score')

review_score    
```

#### Return Title

```python
# Find all review titles
review_titles = []

for lc in link_contents:
    review_titles.append(lc.find('title').text)
    
review_titles
```

### Pull out artist,  album, and publication name

```python
rt_list = []

for rt in review_titles:
    rt_list.append(rt.replace(': ',' | ', 1)) # replace : with |

titles_split = []

for rtl in rt_list:
    titles_split.append(rtl.split(' | '))

titles_split
```

#### Pull out publication name

```python
publication_name_list = []

for ts in titles_split:
    try:
        publication_name_list.append(ts[2])
    except:
        publication_name_list.append('No publication')

publication_name_list
```

#### Pull out artist

```python
artist_list = []

for ts in titles_split:
    artist_list.append(ts[0])

artist_list
```

#### Pull out album

```python
album_list = []

for ts in titles_split:
    try:
        album_list.append(ts[1])
    except:
        album_list.append('No album')

album_list
```

### Replace 'Album Review'



```python
album_reviews = []

for album in album_list:
    album_reviews.append(album.replace('Album Review','').strip())

album_reviews
```

### Find the genre

```python
# Find the genres in the reviews
genre_list = []

for lc in link_contents:
    try:
        genre_list.append(lc.find(class_='genre-list__item').text)
    except:
        genre_list.append('No genre')
                      
genre_list
```

### Find the summaries in the reviews

```python
summaries_list = []

for lc in link_contents:
    try:
        summaries_list.append(lc.find(class_='review-detail__article-content').find('p').text)
    except:
        summaries_list.append('No summary')
                  
summaries_list
```

### Find the content in each review

```python
contents_list = []

for lc in link_contents:
    try:
        contents_list.append(re.sub(r'Back to home', '', \
                                    (re.sub(r'\s', ' ', lc.find(class_='review-detail__article-content').text))))
    except:
        contents_list.append("No review")
        
# contents_list
```
