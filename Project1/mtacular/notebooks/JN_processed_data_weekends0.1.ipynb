```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

from datetime import datetime as dt
import datetime

import code.get_mta_data as gmd
import code.process_mta_data as pmd
import code.analyze_mta_data as amd
```

```python
mta_station_freqs = amd.get_station_freqs(mta_station_grouped)
```

```python
mta_station_reset = mta_station_freqs.reset_index()
mta_station_reset
```

```python
mta_station_reset_weekend = mta_station_reset[mta_station_reset.DAYNAME.isin(['Saturday','Sunday'])]
```

```python
mta_station_reset_weekend.tail()
```

```python
mta_mean_across_weekends = amd.mean_weekend_rankings(mta_station_freqs)
mta_mean_across_weekends
```

```python
mta_mean_across_weekends_top = mta_mean_across_weekends.head(10)
mta_mean_across_weekends_top
```

```python
mta_mean_across_weekends_top.columns
```

```python
# bar plot with top 5 or 10 from weekday and weekend, overlay later

plt.plot(mta_mean_across_weekends_top)
plt.title('MTA MEAN ACROSS WEEKENDS',fontsize = 22,weight='bold',color='blue');
```

```python
ax = mta_mean_across_weekends_top.plot(kind='bar', figsize=(15, 10), legend=True, fontsize=22)
plt.title('MTA MEAN ACROSS WEEKENDS',fontsize = 40,weight='bold',color='blue');
ax.set_xlabel("", fontsize=22, color='black', labelpad=10)
ax.set_ylabel("INS", fontsize=22)

plt.savefig('MTA Mean AX Weekends.svg')
```

```python
x = (mta_station_freqs
     .groupby(['DAY','DAYNAME'])
     .agg({'INS':'sum', 'OUTS':'sum'})
     .reset_index()
     .sort_values(['DAY'],ascending=False)
     ['DAYNAME']
    )
y = (mta_station_freqs
     .groupby('DAY')
     .agg({'INS':'sum', 'OUTS':'sum'})
     .reset_index()
     .sort_values(['DAY'],ascending=False)
     ['INS']/1000000
    )
plt.barh(x,y)
plt.xticks(rotation='vertical')
plt.title('Average Daily MTA Entries')
plt.xlabel('Entries in Millions')

plt.savefig('Avg. Daily MTA Entries All Days.svg')
```

```python
x = (mta_station_reset_weekend
     .groupby(['DAY','DAYNAME'])
     .agg({'INS':'sum', 'OUTS':'sum'})
     .reset_index()
     .sort_values(['DAY'],ascending=False)
     ['DAYNAME']
    )
y = (mta_station_reset_weekend
     .groupby('DAY')
     .agg({'INS':'sum', 'OUTS':'sum'})
     .reset_index()
     .sort_values(['DAY'],ascending=False)
     ['INS']/1000000
    )
plt.barh(x,y)
plt.xticks(rotation='vertical')
plt.yticks(rotation='vertical')
plt.title('Average Daily MTA Entries by Weekend')
plt.xlabel('Entries in Millions');
```
