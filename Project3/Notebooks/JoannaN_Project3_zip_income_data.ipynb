# Import Modules

```python
import pandas as pd
```

```python
from bs4 import BeautifulSoup
import requests
```

# Read in data with BeautifulSoup

```python
# income_zip = pd.read_html('http://zipatlas.com/us/il/chicago/zip-code-comparison/median-household-income.htm')

url = 'http://zipatlas.com/us/il/chicago/zip-code-comparison/median-household-income.htm' 

response = requests.get(url)
```

```python
response.status_code  #200 = success!
```

```python
page = response.text
```

```python
soup = BeautifulSoup(page, "lxml")
```

```python
print(soup.prettify())
```

```python
table = soup.find('table')
table

# yay tables!|
```

# Read tables from html page into a dataframe


```python
income_zip_df = pd.read_html(str(table))[10]
```

```python
income_zip_df.head()
```

```python
income_zip_df.columns
```

## Drop columns I don't need:
- #, Location, City, National Rank

```python
income_zip_df_2 = income_zip_df.drop(columns=[0, 2, 3, 6])
```

```python
income_zip_df_2.head()
```

```python
income_zip_df_2.columns
```

## Change column names to something more useful

```python
income_zip_df_3 = income_zip_df_2.rename\
(columns = {1:'zip', 4: 'population', 5: 'income'}) 
```

```python
income_zip_df_3.head()
```

```python
income_zip_df_3.tail()
```

## Drop first row

```python
income_zip_df_4 = income_zip_df_3.drop([0])
```

## Drop last row

```python
income_zip_df_4 = income_zip_df_4.drop([55])
```

## Convert objects to ints

```python
income_zip_df_4.dtypes
```

```python
income_zip_df_4["zip"]= income_zip_df_4["zip"].astype(int) 
income_zip_df_4["population"]= income_zip_df_4["population"].astype(int) 
```

```python
income_zip_df_4.dtypes
```

#### Convert income from $ to int

```python
income_zip_df_4.columns[2:]
```

```python
income_zip_df_4['income'].head()
```

```python
income_zip_df_4['income'] = income_zip_df_4\
[income_zip_df_4.columns[2:]].replace('[\$,]', '', regex=True).astype(float)
```

```python
income_zip_df_4.dtypes
```
